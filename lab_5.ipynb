```python
import matplotlib.pyplot as plt
import numpy as np
w= np.array([10, 0.2, -0.75])
learning_rate = 0.05
inputs = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
labels = np.array([0, 0, 0, 1])
epochs = 1000
total=[]
count=0
for epoch in range(epochs):
    total_error=0
    square_error=0
    for i in range(len(inputs)):
        input_vector = np.insert(inputs[i], 0, 1) 
        output = np.dot(input_vector, w) > 0 
        error = labels[i] - output
        total_error=total_error+error
        square_error=total_error+error*error
        w = w + learning_rate * error * input_vector
    count=count+1
    if abs(total_error) <0.002:
        break
    total.append(square_error)
print(w[0],w[1],w[2])
print(len(total))
plt.plot(np.arange(0,count-1),total)
plt.title("epochs vs sum square error")
plt.xlabel("epoch")
plt.ylabel("sum square error")
```
