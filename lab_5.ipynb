```python
import matplotlib.pyplot as plt
import numpy as np
w= np.array([10, 0.2, -0.75])
learning_rate = 0.05
inputs = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
labels = np.array([0, 0, 0, 1])
epochs = 1000
total=[]
square=[]
count=0
for epoch in range(epochs):
    total_error=0
    square_error=0
    for i in range(len(inputs)):
        input_vector = np.insert(inputs[i], 0, 1) 
        output = np.dot(input_vector, w) > 0 
        error = labels[i] - output
        total_error=total_error+error
        square_error=square_error+(error**2)
        w = w + learning_rate * error * input_vector
    count=count+1
    total.append(total_error)
    square.append(square_error)
    if abs(total_error) <0.002:
        break
print(w[0],w[1],w[2])
print(len(total))
plt.plot(np.arange(0,count),total,color='blue')
plt.title("epochs vs total error")
plt.xlabel("epoch")
plt.ylabel("total error")
```

```python
plt.plot(np.arange(0,count),square,color='red')
plt.title("epochs vs sum square error")
plt.xlabel("epoch")
plt.ylabel("sum square error")
```
